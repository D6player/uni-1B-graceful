```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 5e-7, int(1e6))
v1 = 3*0.9558*np.cos(2*np.pi*10.137e6*t)
v2 = 3 + 3*np.cos(2*np.pi*10.137e6*t)
```

```python
fig = plt.figure(figsize=(15, 3))

plt.plot(t, v1, label="Sortida")
plt.plot(t, v2, label="Entrada")
plt.title("Tensió d\'entrada i sortida en funció del temps")
plt.xlabel("Temps (s)")
plt.ylabel("Tensió (V)")
plt.xticks([i/2 * 1e-7 for i in range(11)])
plt.grid()
plt.legend()

plt.savefig("sortida1.pdf", bbox_inches = 'tight')
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 3))

axs[0].stem([-10.137e6, 0, 10.137e6], [1.5*0.9558, 0, 1.5*0.9558])
axs[0].set_yticks([i/2 for i in range(8)])
axs[0].set_xlabel("Freqüència (Hz)")
axs[0].set_ylabel("Amplitud")
axs[0].set_title("Mòdul de l\'espectre de la sortida")
axs[0].grid()

axs[1].stem([-10.137e6, 0, 10.137e6], [1.5, 3, 1.5], "C1")
axs[1].set_yticks([i/2 for i in range(8)])
axs[1].set_xlabel("Freqüència (Hz)")
axs[1].set_ylabel("Amplitud")
axs[1].set_title("Mòdul de l\'espectre de l\'entrada")
axs[1].grid()

plt.savefig("sortida2.pdf", bbox_inches = 'tight')
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 3))

axs[0].plot([-10.137e6, 0, 10.137e6], [0, 0, 0])
axs[0].set_title("Fase de l\'espectre de la sortida")
axs[0].set_xlabel("Freqüència (Hz)")
axs[0].set_ylabel("Fase (deg)")
axs[0].grid()

axs[1].plot([-10.137e6, 0, 10.137e6], [0, 0, 0], "C1")
axs[1].set_title("Fase de l\'espectre de l\'entrada")
axs[1].set_ylabel("Fase (deg)")
axs[1].set_xlabel("Freqüència (Hz)")
axs[1].grid()

plt.savefig("sortida3.pdf", bbox_inches = 'tight')
```
