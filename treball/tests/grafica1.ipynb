```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Extraccio de les dades

with open("Mesures.s2p", "r") as f:
    raw_data = f.read().split("\n")[1:-1]
    raw_data = [line.split() for line in raw_data]

freqs = np.array([ float(line[0]) for line in raw_data ])
phasors = np.array([ float(line[3]) + 1j*float(line[4]) for line in raw_data ])
```

```python
# Obtencio dels moduls i les fases

modules = np.abs(phasors)
phases = np.angle(phasors, deg=True)
```

```python
# Obtencio del guany

gains = 20 * np.log10(modules)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
#fig.suptitle("54894677W")
axs[0].plot(freqs, gains)
axs[0].set_title("Guany en funció de freqüència")
axs[0].set_xlabel("Freqüència (Hz)")
axs[0].set_ylabel("Guany (dB)")
axs[0].grid()
axs[1].plot(freqs, phases)
axs[1].set_title("Fase en funció de freqüència")
axs[1].set_xlabel("Freqüència (Hz)")
axs[1].set_ylabel("Fase (deg)")
axs[1].grid()

fig.savefig("resposta_frequencial.pdf", bbox_inches = 'tight')
```
