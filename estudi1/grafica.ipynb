```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("grafica.dat", "r") as f:
    data = f.read().split()[2:]
    data = [float(i) for i in data]
```

```python
n = len(data)

x = [data[i*2] for i in range(n//2)]
y = [data[i*2 + 1] for i in range(n//2)]

plt.grid()
plt.ylabel("Vo (Volt)")
plt.xlabel("Vi (Volt)")
plt.title("Característica de sortida")
plt.plot(x, y)
plt.savefig("grafica.png")
```
